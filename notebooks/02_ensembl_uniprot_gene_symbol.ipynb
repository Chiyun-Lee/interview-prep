```python
import pandas as pd

df_full = pd.read_csv("../data/HUMAN_9606_idmapping.dat.gz", sep="\t", header=None,
                      names=["uniprot_id", "db", "id"])

ensembl = df_full.loc[df_full.db == "Ensembl", ["uniprot_id", "id"]].rename(columns={"id": "ensembl_gene_id"})
symbols = df_full.loc[df_full.db == "Gene_Name", ["uniprot_id", "id"]].rename(columns={"id": "gene_symbol"})

final = ensembl.merge(symbols, on="uniprot_id", how="outer").dropna()

final.ensembl_gene_id = final.ensembl_gene_id.str.split(".").str[0]

final = final.loc[final.uniprot_id.str.len() == 6].drop_duplicates()
```

```python
final.to_json("../data/gene_mapping.json", orient="records", indent=1)
```
